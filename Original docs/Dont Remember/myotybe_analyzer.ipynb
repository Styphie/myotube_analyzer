# Myotube Analyzer Project

This project is designed to analyze fluorescence microscopy images of myotubes to quantify various aspects of muscle cell fusion and differentiation. The analysis pipeline takes raw images as input and produces quantitative data on myotube formation, including fusion index, number of nuclei per myotube, and other relevant metrics.

## Project Overview

The Myotube Analyzer project aims to automate the process of analyzing myotube images, which is often a time-consuming and manual task. By leveraging image processing and analysis techniques, this project provides a streamlined workflow for researchers studying myogenesis.

The core of the project involves identifying and quantifying myotubes and their nuclei from fluorescence microscopy images. The analysis is based on specific staining techniques that label cell nuclei (e.g., with DAPI) and myosin heavy chain (MyHC), a marker for differentiated muscle cells.



## Data

The project utilizes two main types of data:

1.
Fluorescence Microscopy Images: These are the primary input for the analysis. The images are expected to be in a standard format (e.g., PNG, TIFF) and should show myotubes stained for nuclei and MyHC.

2.
Experimental Data: An Excel file contains the experimental details and manually analyzed data. This file includes multiple sheets, each corresponding to different passages of myoblasts. The columns in the Excel file provide quantitative metrics for each well, such as:

• `Passage #`: The passage number of the myoblasts.

• `Well #`: The well number in the experimental plate.

• `# of Nuclei`: The total number of nuclei in the image.

• `# of MyHC Positive Nuclei`: The number of nuclei within MyHC-positive myotubes.

• `% of MyHC Positive Nucei`: The percentage of nuclei that are MyHC-positive.

• `# of Nucei that have Fused (MyHC positive-2 or more)`: The number of nuclei in myotubes containing two or more nuclei.

• `# of Myotubes`: The total number of myotubes in the image.

• `Average # of Nuclei per Myotube`: The average number of nuclei per myotube.

• `Fusion Index (%)`: The percentage of nuclei that have fused to form myotubes
